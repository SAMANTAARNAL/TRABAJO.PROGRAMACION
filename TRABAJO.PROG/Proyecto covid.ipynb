DESCRIPCIÓN DE LOS DATOS:

Para el desarrollo del proyecto se han seleccionado una serie de datos en formato .csv, proporcionados en abierto por el Ministerio de Sanidad, a través del Centro de Coordinación de Alertas y Emergencias Sanitarias, respecto a la evolución de la afección por Coronavirus en la población española. La base de datos recoje los datos diarios de casos de COVID-19 diagnosticados, los que han tenido que ser hospitalizados, los que han sido trasladados a UCI´s, y los casos de fallecimiento; todos ellos, diferenciados por comunidad autónoma.

Se ha optado por hacer las diferentes gráficas a partir de los datos recopilados desde el 1 de enero de 2020 hasta el 8 de diciembre de 2010. Se ha descargado el fichero de datos de entrada la página web del Ministerio de Sanidad (en formato .csv): "https://www.mscbs.gob.es/profesionales/saludPublica/ccayes/alertasActual/nCov/documentos/Datos_Casos_COVID19.csv".
```python
# El primer paso es incluir todas las librerías que se van a utilizar. En nuestro caso, sera numpy, matplotlib y pandas,
# todas ellas necesarias para trabajar con nuestro dataset y crear gráficas


import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 

datos_Covid = pd.read_csv('DatosCasos_COVID19.csv',delimiter=';',encoding = "ISO-8859-1", header = 0) 
datos_Covid.set_index('Fecha', inplace = True, drop = True)  

# Es necesario diferenciar las CC.AA.:
Andalucia = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-AN"]
Aragon = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-AR"]
Asturias = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-AS"]
Canarias = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-CN"]
Cantabria = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-CB"]
Castilla_M = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-CM"]
Castilla_L = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-CL"]
Catalunya = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-CT"]
Extremadura = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-EX"]
Galicia = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-GA"]
Baleares = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-IB"]
Rioja = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-RI"]
Madrid = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-MD"]
Murcia = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-MC"]
Navarra = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-NC"]
Pais_V = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-PV"]
Valenciana = datos_Covid[datos_Covid["CCAA_ISO"] == "ES-VC"]


# Primero, se van a crear gráficas según estado del paciente (contagiado, ingresado o fallecido) a nivel nacional.
# Después, se especificará según comunidad autónoma.

fechas = datos_Covid.index 

lista_fechas = fechas.drop_duplicates().tolist()[::15] 
plt.xticks(np.arange(len(lista_fechas))*(len(fechas.drop_duplicates())/len(lista_fechas)),lista_fechas,rotation=90) 
```

```python
plt.plot(datos_Covid.Casos_Diagnosticados, label = 'Casos diagnosticados')
plt.legend()
plt.xlabel('Fecha')
plt.title('España')
plt.ylabel('Número de contagiados')
plt.show()
```

```python
plt.xticks(np.arange(len(lista_fechas))*(len(fechas.drop_duplicates())/len(lista_fechas)),lista_fechas,rotation=90)
plt.plot(datos_Covid.Hospitalizados, label = 'Ingresados', color = 'r')
plt.legend()
plt.title('España')
plt.xlabel('Fecha')
plt.ylabel('Número de ingresados')
plt.show()
```
